#TABELA DE AMOSTRAGEM(MÉTODO ANTIGO)

```python
largura_maxima = 2

def formatar_tabela(numeros, num_por_linha = 20, largura = largura_maxima):
  for i in range(0, len(numeros), num_por_linha):
    linha = numeros[i:i+num_por_linha]
    print(*[str(num).rjust(largura)for num in linha])

formatar_tabela(numeros_aleatorios)
```

```python
def amostragem_aleatoria_simples(dataset, amostras):
  return dataset.sample(n = amostras, random_state=1)
```

```python
df_amostra_aleatoria_simples = amostragem_aleatoria_simples(dataset, 100)
df_amostra_aleatoria_simples.shape
```
